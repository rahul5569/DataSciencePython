```python
WFC = web.DataReader("WFC", "av-daily", start=datetime(2006, 1, 1),end=datetime(2022, 1, 1),api_key=os.getenv('ALPHAVANTAGE_API_KEY'))
```

Creating the data frame

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],keys = tickers,axis = 1)
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

## EDA
Plotting the stock data.


```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# Optional Plotly Method Imports
import plotly
import cufflinks as cf
cf.go_offline()

```

```python
bank_stocks[:].xs(key = 'close',axis = 1,level = 'Stock Info').iplot(kind='spread')
```

What I noticed was there was huge spike in pct change for Goldman Sachs bank from march 2020 to april 2021(approx, used cufflinks to check). Since I knew to calculate only returns using pct_change() hence I am using this for now.

```python
returns = pd.DataFrame()
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['close'].pct_change(periods = 308) 

## Why I have taken 308 period? Don't know exactly XD. I wanted to get trend but lower periods did not gave me trend hence,
## I took a random higher value.
```

Taking only returns of top 500 as I wanted max retruns.

```python
retrunTop500 = pd.DataFrame()
for tic in tickers:
    retrunTop500[tic + ' Return'] = returns[tic + ' Return'].nlargest(500)
```

Plotting the returns

```python
retrunTop500.sort_index().iplot(kind = 'spread')
```
