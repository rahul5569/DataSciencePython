**Plotting of Linear equations**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000)
```

```python
d_r = 2.5 * t
```

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title("Graph of sheriff catching banking robber")
plt.xlabel("Time in minutes")
plt.ylabel("Distance in km")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
plt.plot(t, d_r, c = "blue")
plt.plot(t, d_s, c = "orange")
plt.axvline(x=30, color = "Red", linestyle = "--")  
plt.axhline(y = 75 , color = "red", linestyle = "--")
```

Now I never understood as to why we are doing t-5 in order calculate distance so I decided to plot it. By putting t+5.  

```python
d_s = 3 * (t+5)
```

```python
fig, ax = plt.subplots()
plt.title("Graph of sheriff catching banking robber")
plt.xlabel("Time in minutes")
plt.ylabel("Distance in km")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
plt.plot(t, d_r, c = "blue")
plt.plot(t, d_s, c = "orange")
```
